```python
import mysql.connector

# Σύνδεση στη βάση
connection = mysql.connector.connect(
    user="root",
    password="Giannis_Ntz_01",
    host="localhost",
    database="littleLemonDb"
)

cursor = connection.cursor(buffered=True)

# SQL query ως συμβολοσειρά
query = """
SELECT 
    CONCAT(Customers.firstName, ' ', Customers.lastName) AS FullName,
    Customers.email,
    Customers.phone,
    Orders.totalAmount
FROM 
    Customers
INNER JOIN 
    Orders
ON 
    Customers.customerID = Orders.customerID
WHERE 
    Orders.totalAmount > 60;
"""

# Εκτέλεση query
cursor.execute(query)

# Λήψη αποτελεσμάτων
results = cursor.fetchall()

# Προβολή αποτελεσμάτων
for row in results:
    print(f"Όνομα: {row[0]}, Email: {row[1]}, Τηλέφωνο: {row[2]}, Ποσό: ${row[3]}")

# Κλείσιμο cursor και σύνδεσης
cursor.close()
connection.close()
```
